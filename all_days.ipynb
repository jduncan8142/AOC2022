The jungle must be too overgrown and difficult to navigate in vehicles or access from the air; the Elves' expedition traditionally goes on foot. As your boats approach land, the Elves begin taking inventory of their supplies. One important consideration is food - in particular, the number of Calories each Elf is carrying (your puzzle input).

```

9000

10000
```

This list represents the Calories of the food carried by five Elves:

The first Elf is carrying food with 1000, 2000, and 3000 Calories, a total of 6000 Calories.
The second Elf is carrying one food item with 4000 Calories.
The third Elf is carrying food with 5000 and 6000 Calories, a total of 11000 Calories.
The fourth Elf is carrying food with 7000, 8000, and 9000 Calories, a total of 24000 Calories.
The fifth Elf is carrying one food item with 10000 Calories.

1. Find the Elf carrying the most Calories. 
2. How many total Calories is that Elf carrying? => Answer: _______


````python
from pathlib import Path

data = {}
raw_data = None

with open(Path(r"C:\Users\duncan\Downloads\day1.txt"), "r") as f:
    raw_data = f.readlines()

elf_data = []
elf_index = 0
for line in raw_data:
    if line != "\n":
        elf_data.append(int(line.strip(" \n\r")))
    else:
        elf_index += 1
        data[elf_index] = {"data": elf_data, "sum": sum(elf_data)}
        elf_data = []

calories = None
top_elf = None
for index, value in data.items():
    if calories is None:
        calories = value["sum"]
        top_elf = index
    elif value["sum"] > calories:
        calories = value["sum"]
        top_elf = index

print(f"Elf {top_elf} has the most food with {calories} calories total.")
````

# Day 2: Rock Paper Scissors
The Elves begin to set up camp on the beach. To decide whose tent gets to be closest to the snack storage, a giant Rock Paper Scissors tournament is already in progress.

Rock Paper Scissors is a game between two players. Each game contains many rounds; in each round, the players each simultaneously choose one of Rock, Paper, or Scissors using a hand shape. Then, a winner for that round is selected: Rock defeats Scissors, Scissors defeats Paper, and Paper defeats Rock. If both players choose the same shape, the round instead ends in a draw.

Appreciative of your help yesterday, one Elf gives you an encrypted strategy guide (your puzzle input) that they say will be sure to help you win. "The first column is what your opponent is going to play: A for Rock, B for Paper, and C for Scissors. The second column--" Suddenly, the Elf is called away to help with someone's tent.

The second column, you reason, must be what you should play in response: X for Rock, Y for Paper, and Z for Scissors. Winning every time would be suspicious, so the responses must have been carefully chosen.

The winner of the whole tournament is the player with the highest score. Your total score is the sum of your scores for each round. The score for a single round is the score for the shape you selected (1 for Rock, 2 for Paper, and 3 for Scissors) plus the score for the outcome of the round (0 if you lost, 3 if the round was a draw, and 6 if you won).

Since you can't be sure if the Elf is trying to help you or trick you, you should calculate the score you would get if you were to follow the strategy guide.

For example, suppose you were given the following strategy guide:

```
A Y
B X
C Z
```

This strategy guide predicts and recommends the following:

- In the first round, your opponent will choose Rock (A), and you should choose Paper (Y). This ends in a win for you with a score of 8 (2 because you chose Paper + 6 because you won).
- In the second round, your opponent will choose Paper (B), and you should choose Rock (X). This ends in a loss for you with a score of 1 (1 + 0).
- The third round is a draw with both players choosing Scissors, giving you a score of 3 + 3 = 6.

In this example, if you were to follow the strategy guide, you would get a total score of 15 (8 + 1 + 6).

What would your total score be if everything goes exactly according to your strategy guide? => Answer: _________

````python
from enum import Enum

strategy_guide = [("A", "Y"), ("B", "X"), ("C", "Z")]

class Rules(Enum):
    L = [("ROCK", "SCISSORS"), ("PAPER", "ROCK"), ("SCISSORS", "PAPER")]
    W = [("SCISSORS", "ROCK"), ("ROCK", "PAPER"), ("PAPER", "SCISSORS")]

class Points(Enum):
    ROCK = 1
    PAPER = 2
    SCISSORS = 3

class Score(Enum):
    L = 0
    D = 3
    W = 6

class Me(Enum):
    X = "ROCK"
    Y = "PAPER"
    Z = "SCISSORS"

class Player(Enum):
    A = "ROCK"
    B = "PAPER"
    C = "SCISSORS"

def calculate_outcome(selections: tuple[str]) -> Score:
    if selections[0] == selections[1]:
        return "D"
    if selections in Rules.L.value:
        return "L"
    if selections in Rules.W.value:
        return "W"
    

def calculate_strategy_total_score(strategy: list[tuple]) -> int:
    scores = []
    for i, round in enumerate(strategy):
        __player = getattr(Player, round[0]).value
        __me = getattr(Me, round[1]).value
        __selections = (__player, __me)
        __outcome = calculate_outcome(__selections)
        __score = getattr(Score, __outcome).value + getattr(Points, __me).value
        scores.append(__score)
        # print(f"Player's: {__player} vs My: {__me}, You: {__outcome}, Score: {__score}")
    return sum(scores)


raw_data = None
data = []
with open(r"C:\Users\duncan\Downloads\day2.txt") as f:
    raw_data = f.readlines()
    
    for line in raw_data:
        __line = line.strip("\n")
        data.append(tuple(__line.split(" ")))

total_score = calculate_strategy_total_score(strategy=data)
print(total_score)
````

# Part Two

The Elf finishes helping with the tent and sneaks back over to you. "Anyway, the second column says how the round needs to end: X means you need to lose, Y means you need to end the round in a draw, and Z means you need to win. Good luck!"

The total score is still calculated in the same way, but now you need to figure out what shape to choose so the round ends as indicated. The example above now goes like this:

- In the first round, your opponent will choose Rock (A), and you need the round to end in a draw (Y), so you also choose Rock. This gives you a score of 1 + 3 = 4.
- In the second round, your opponent will choose Paper (B), and you choose Rock so you lose (X) with a score of 1 + 0 = 1.
- In the third round, you will defeat your opponent's Scissors with Rock for a score of 1 + 6 = 7.

Now that you're correctly decrypting the ultra top secret strategy guide, you would get a total score of 12.

Following the Elf's instructions for the second column, what would your total score be if everything goes exactly according to your strategy guide? => Answer: ___________

````python
from enum import Enum

strategy_guide = [("A", "Y"), ("B", "X"), ("C", "Z")]

class Rules(Enum):
    L = [("ROCK", "SCISSORS"), ("PAPER", "ROCK"), ("SCISSORS", "PAPER")]
    W = [("SCISSORS", "ROCK"), ("ROCK", "PAPER"), ("PAPER", "SCISSORS")]

class Points(Enum):
    ROCK = 1
    PAPER = 2
    SCISSORS = 3

class Score(Enum):
    L = 0
    D = 3
    W = 6

class Me(Enum):
    X = "L"
    Y = "D"
    Z = "W"

class Player(Enum):
    A = "ROCK"
    B = "PAPER"
    C = "SCISSORS"

def new_calculate_outcome(selections: tuple[str]) -> Score:
    if selections[1] == "D":
        return selections[0]
    else:
        responses = getattr(Rules, selections[1]).value
        for i in responses:
            if i[0] == selections[0]:
                return i[1]
    

def calculate_strategy_total_score(strategy: list[tuple]) -> int:
    scores = []
    for i, round in enumerate(strategy):
        __player = getattr(Player, round[0]).value
        __outcome = getattr(Me, round[1]).value
        __selections = (__player, __outcome)
        __me = new_calculate_outcome(__selections)
        __score = getattr(Score, __outcome).value + getattr(Points, __me).value
        scores.append(__score)
        # print(f"Player's: {__player} vs My: {__me}, You: {__outcome}, Score: {__score}")
    return sum(scores)


raw_data = None
data = []
with open(r"C:\Users\duncan\Downloads\day2.txt") as f:
    raw_data = f.readlines()
    
    for line in raw_data:
        __line = line.strip("\n")
        data.append(tuple(__line.split(" ")))

total_score = calculate_strategy_total_score(strategy=data)
print(total_score)
````

Space needs to be cleared before the last supplies can be unloaded from the ships, and so several Elves have been assigned the job of cleaning up sections of the camp. Every section has a unique ID number, and each Elf is assigned a range of section IDs.

In how many assignment pairs does one range fully contain the other? => Answer: ___________

````python
data = []
raw_data = None
count = 0
with open(r"C:\Users\duncan\Downloads\day4.txt", "r") as f:
    raw_data = [i.split(",") for i in [x.strip("\n") for x in f.readlines()]]

for i in raw_data:
    pairs = []
    for j in i:
        t = j.split("-")
        tmp = [x for x in range(int(t[0]), int(t[1])+1)]
        pairs.append(tmp)
    data.append(pairs)

for i in data:
    r1 = set(i[0])
    r2 = set(i[1])
    if r1.issubset(r2) or r2.issubset(r1):
        count += 1

print(f"Count: {count}")
````

# Day 6: Tuning Trouble

The preparations are finally complete; you and the Elves leave camp on foot and begin to make your way toward the star fruit grove.

As you move through the dense undergrowth, one of the Elves gives you a handheld device. He says that it has many fancy features, but the most important one to set up right now is the communication system.

However, because he's heard you have significant experience dealing with signal-based systems, he convinced the other Elves that it would be okay to give you their one malfunctioning device - surely you'll have no problem fixing it.

As if inspired by comedic timing, the device emits a few colorful sparks.

To be able to communicate with the Elves, the device needs to lock on to their signal. The signal is a series of seemingly-random characters that the device receives one at a time.

To fix the communication system, you need to add a subroutine to the device that detects a start-of-packet marker in the datastream. In the protocol being used by the Elves, the start of a packet is indicated by a sequence of four characters that are all different.

The device will send your subroutine a datastream buffer (your puzzle input); your subroutine needs to identify the first position where the four most recently received characters were all different. Specifically, it needs to report the number of characters from the beginning of the buffer to the end of the first such four-character marker.

For example, suppose you receive the following datastream buffer:

`mjqjpqmgbljsphdztnvjfqwrcgsmlb`

After the first three characters (`mjq`) have been received, there haven't been enough characters received yet to find the marker. The first time a marker could occur is after the fourth character is received, making the most recent four characters `mjqj`. Because `j` is repeated, this isn't a marker.

The first time a marker appears is after the seventh character arrives. Once it does, the last four characters received are `jpqm`, which are all different. In this case, your subroutine should report the value `7`, because the first start-of-packet marker is complete after 7 characters have been processed.

Here are a few more examples:

- `bvwbjplbgvbhsrlpgdmjqwftvncz`: first marker after character `5`
- `nppdvjthqldpwncqszvftbrmjlhg`: first marker after character `6`
- `nznrnfrfntjfmvfwmzdfjlvtqnbhcprsg`: first marker after character `10`
- `zcfzfwzzqfrljwzlrfnpqdbhtmscgvjw`: first marker after character `11`

How many characters need to be processed before the first start-of-packet marker is detected? => Answer: ___________


````python
def is_marker_valid(marker: list[str]) -> bool:
    __marker = marker
    while len(__marker) > 0:
        char = __marker.pop(0)
        if char in __marker:
            return False
        else:
            continue
    return True


def parse_datastream(datastream: str, rule: int) -> int:
    packet_marker: list[str] = []
    for index, char in enumerate([i for i in datastream]):
        packet_marker.append(char)  # Add current char to marker
        # If length of marker is greater than 4 then trim oldest char from marker to maintain length 4.
        if len(packet_marker) > rule:
            _ = packet_marker.pop(0)
        # Confirm marker length is 4 and check if marker is valid marker. 
        if len(packet_marker) == rule and is_marker_valid(marker=packet_marker):
            return index + 1
    return None


with open(r"C:\Users\jason\Downloads\day6.txt", "r") as f:
    print(f"Answer = {parse_datastream(f.read(), 4)}")
````

````python
from dataclasses import dataclass
from typing import Optional
import re


@dataclass
class Command:
    raw_text: str
    cmd: Optional[str|None] = None
    arguments: Optional[list[str]|None] = None
    error: Optional[str|None] = None

    def __post_init__(self) -> None:
        try:
            __txt = self.raw_text.split(" ")
        except Exception as err:
            self.error = f"ERROR: {err}"
        try:
            if __txt[0] != "$":
                self.error = f"ERROR: Command lines must start with '$', this line starts with {__txt[0]}"
        except Exception as err:
            self.error = f"ERROR: {err}"
        try:
            self.cmd = __txt[1].lower()
            if len(__txt) > 2:
                self.arguments = [x.lower() for x in __txt[2:]]
        except Exception as err:
            self.error = f"ERROR: {err}"

@dataclass
class File:
    raw_text: str
    name: Optional[str|None] = None
    directory: Optional[str|None] = None
    size: Optional[int|None] = None
    error: Optional[str|None] = None

    def __post_init__(self) -> None:
        try:
            try:
                __txt = self.raw_text.split(" ")
            except Exception as err:
                self.error = f"ERROR: {err}"
            try:
                __r = re.match(r"\d*", __txt[0])
                if __r is None:
                    self.error = f"ERROR: File output must start with numeric size, this line starts with {__txt[0]}"
            except Exception as err:
                self.error = f"ERROR: {err}"
            try:
                self.name = __txt[1]
                self.size = int(__txt[0])
            except Exception as err:
                self.error = f"ERROR: {err}"
        except Exception as err:
            self.error = f"ERROR: {err}"

@dataclass
class Directory:
    name: str
    children: Optional[list|None] = None
    files: Optional[list|None] = None
    parent: Optional[str|None] = None

    @property
    def directory_size(self) -> int:
        total = 0
        for file in self.files:
            total += file.size
        for d in self.children:
            tmp = d.directory_size
            total += tmp
        return total


def print_file(f: File, lvl: int) -> None:
    tab = str("\t") * lvl
    print(f"{tab}File: {f.name}, Size: {f.size}")


def print_dir(d: Directory, lvl: int) -> None:
    tab = str("\t") * lvl
    print(f"{tab}Dir: {d.name}, Size: {d.directory_size}")
    for c in d.children:
        print_dir(c, lvl+1)
    for f in d.files:
        print_file(f, lvl+1)


def parse_terminal_output(output: str) -> list:
    __parsed_commands: list[Command] = []
    __current_dir: Directory = None
    __current_cmd: Command = None
    __ls_output_flag: bool = False
    __output: list = output.split("\n")
    __previous_cmd: Command = None
    __root = Directory(name="/", children=[], files=[], parent=None)
    for index, line in enumerate(__output):
        if line.startswith("$"):
            __previous_cmd = __current_cmd
            __current_cmd = Command(raw_text=line)
            if __current_cmd.error is None:
                __parsed_commands.append(__current_cmd)
            if __current_cmd.cmd == "cd":
                __ls_output_flag = False
                if len(__current_cmd.arguments) > 0:
                    if __current_cmd.arguments[0] == "/":
                        __current_dir = __root
                    elif __current_cmd.arguments[0] == "..":
                        if __current_dir.parent is not None and isinstance(__current_dir.parent, Directory):
                            __current_dir = __current_dir.parent
                        else:
                            __current_dir = __root
                    else:
                        if __current_cmd.arguments[0] in [x.name for x in __current_dir.children]:
                            __current_dir = [d for d in __current_dir.children if __current_cmd.arguments[0] == d.name][0]
                        else:
                            __current_dir = Directory(name=__current_cmd.arguments[0], children=[], files=[], parent=__current_dir)
            elif __current_cmd.cmd == "ls":
                __ls_output_flag = True
        elif __ls_output_flag:
            if line.startswith("dir"):
                __current_dir.children.append(Directory(name=line.split(" ")[1], children=[], files=[], parent=__current_dir))
            else:
                __file = File(raw_text=line, directory=__current_dir)
                if __file.name is not None:
                    __current_dir.files.append(__file)
    return __root

r = parse_terminal_output(data)
# print(f"Answer = {r}")
````
